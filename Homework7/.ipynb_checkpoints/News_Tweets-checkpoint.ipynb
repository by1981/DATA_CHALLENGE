```python
# Dependencies
import tweepy
import json
import numpy as np
from datetime import datetime
from config import consumer_key, consumer_secret, access_token, access_token_secret
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```python
# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
# Target User: 
target_user1 ="@CNN"
target_user2="@BBC"
target_user3="@CBS"
target_user4="@nytimes"
target_user5="@FoxNews"
```

```python
# Get all tweets from home feed
tweet_text1 = []
tweet_text2 = []
tweet_text3 = []
tweet_text4 = []
tweet_text5 = []
compound_list1 = []
positive_list1 = []
negative_list1 = []
neutral_list1 = []
compound_list2 = []
positive_list2 = []
negative_list2 = []
neutral_list2 = []
compound_list3 = []
positive_list3 = []
negative_list3 = []
neutral_list3 = []
compound_list4 = []
positive_list4 = []
negative_list4 = []
neutral_list4 = []
compound_list5 = []
positive_list5 = []
negative_list5 = []
neutral_list5 = []

for status in tweepy.Cursor(api.user_timeline, id=target_user1,tweet_mode="extended").items(100):
        tweet=status._json
        result=analyzer.polarity_scores(tweet['full_text'])
        tweet_text1.append(tweet['full_text'])
        compound_list1.append(result['compound'])
        positive_list1.append(result['pos'])
        negative_list1.append(result['neg'])
        neutral_list1.append(result['neu'])

for status in tweepy.Cursor(api.user_timeline, id=target_user2,tweet_mode="extended").items(100):
        tweet=status._json
        result=analyzer.polarity_scores(tweet['full_text'])
        tweet_text2.append(tweet['full_text'])
        compound_list2.append(result['compound'])
        positive_list2.append(result['pos'])
        negative_list2.append(result['neg'])
        neutral_list2.append(result['neu'])
            
for status in tweepy.Cursor(api.user_timeline, id=target_user3,tweet_mode="extended").items(100):
        tweet=status._json
        result=analyzer.polarity_scores(tweet['full_text'])
        tweet_text3.append(tweet['full_text'])
        compound_list3.append(result['compound'])
        positive_list3.append(result['pos'])
        negative_list3.append(result['neg'])
        neutral_list3.append(result['neu'])
        
for status in tweepy.Cursor(api.user_timeline, id=target_user4,tweet_mode="extended").items(100):
        tweet=status._json
        result=analyzer.polarity_scores(tweet['full_text'])
        tweet_text4.append(tweet['full_text'])
        compound_list4.append(result['compound'])
        positive_list4.append(result['pos'])
        negative_list4.append(result['neg'])
        neutral_list4.append(result['neu'])
       
for status in tweepy.Cursor(api.user_timeline, id=target_user5,tweet_mode="extended").items(100):
        
        tweet=status._json
        result=analyzer.polarity_scores(tweet['full_text'])
        tweet_text5.append(tweet['full_text'])
        compound_list5.append(result['compound'])
        positive_list5.append(result['pos'])
        negative_list5.append(result['neg'])
        neutral_list5.append(result['neu'])
        


      
print(len(compound_list5))
print(len(tweet_text2))
print(len(tweet_text3))
print(len(tweet_text4))
print(len(tweet_text5))
```

```python
x=range(0,len(tweet_text1))
plt.figure(figsize=(9,7))
plt.scatter(x,compound_list1, s=100 ,color='red',edgecolors="black" )
plt.scatter(x,compound_list2, s=100 ,color='lightblue',edgecolors="black" )
plt.scatter(x,compound_list3, s=100 ,color='green' ,edgecolors="black")
plt.scatter(x,compound_list4, s=100 ,color='yellow' ,edgecolors="black")
plt.scatter(x,compound_list5, s=100 ,color='blue',edgecolors="black" )

now = datetime.now()
now = now.strftime("%Y-%m-%d %H:%M")
plt.title(f"Sentiment Analysis of Tweets ({now}) for {target_user1},{target_user2},{target_user3},{target_user4}, and {target_user5}")
plt.ylabel("Tweet Polarity")
plt.xlabel("Tweets Ago")
#ax.legend()
#plt.legend(target_user1,target_user2,target_user3,target_user4,target_user5)
ax = plt.gca().invert_xaxis()
```
