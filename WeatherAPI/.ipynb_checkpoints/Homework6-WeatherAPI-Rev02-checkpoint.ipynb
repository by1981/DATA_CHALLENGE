```python
import json
import requests
import pandas as pd
from config import api_key
import random
import citipy
import openweathermapy.core as owm
units='imperial'
```

```python

cities_csv="worldcities.csv"
cities_info_df=pd.read_csv(cities_csv)
cities_info_df.head()
```

```python
url='http://api.openweathermap.org/data/2.5/weather?'

query_urllist=[]
lat_random=[]
long_random=[]
city_namelist=[]
country_codelist=[]




for i in range(20):
    lat_random.append(random.choice(cities_info_df["Latitude"]))
    long_random.append(random.choice(cities_info_df["Longitude"]))
    
#query_urllist.append(url+"lat="+lat_random+"lon="+long_random+"&appid"+api_key)
for i in range(20):
    cityname=(citipy.nearest_city(float(lat_random[i]), float(lat_random[i])))
    city_namelist.append(cityname.city_name)
    country_codelist.append(cityname.country_code)

#print(city_namelist)
#print(country_codelist)
import urllib

    
city_dict={ 'City':city_namelist,
            'Country Code':country_codelist,
            'Latitude': lat_random,
            'Longitude': long_random
          }
city_dict_df1=pd.DataFrame(city_dict)
city_dict_df=city_dict_df1.drop_duplicates(["City"])

city_dict_df=city_dict_df.reset_index(drop=True)
print(city_dict_df)
for i in range(len(city_dict_df)):
    query_url=f"{url}appid={api_key}&units={units}&q="
    #print(city_dict_df['City'][i])
    #print(query_url+city_namelist[i])
    query_urllist.append(query_url+city_dict_df['City'][i].replace(" ", "%20"))
    
    print(query_urllist[i])    
```

```python
temp=[]
humid=[]
cloud=[]
wind=[]
lat=[]
country=[]
date=[]
lng=[]
max_temp=[]
city=[]
for i in range(len(city_dict_df)):
#     print(city_dict_df['City'])
    try:
        response = requests.get(query_urllist[i]).json()   
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        country.append(response['sys']['country'])
        city.append(response['name'])
        date.append(response['dt'])
        max_temp.append(response['main']['temp_max'])
        temp.append(response['main']['temp'])
        humid.append(response['main']['humidity'])
        wind.append(response['wind']['speed'])
        cloud.append(response['clouds']['all'])
    except:
        print("error happens at iteration", i)
        pass
city
```

```python
weather_dict={'City':city,
            "Lat": lat,
            "Temp": temp,
            "Humidity": humid,
            "Wind Speed": wind,
            "Cloudiness": cloud,
            "Lng":lng,
            "Date":date,
            "Country":country,
            "Max Temp":max_temp
             }
weather_data = pd.DataFrame(weather_dict)
weather_data
```
