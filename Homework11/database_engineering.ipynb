```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
import pymysql
pymysql.install_as_MySQLdb()
import pandas as pd
import os
import numpy as np
import datetime as dt
from sqlalchemy import Column, Integer, String, Float, ForeignKey
```

```python
engine = create_engine("sqlite:///hawaii_2.sqlite")

Base=declarative_base()
session=Session(bind=engine)
conn=engine.connect()

# Create classes 
class Measurement(Base):
    __tablename__ = "measurement"
    id = Column(Integer, primary_key=True)
    station = Column(String)
    date = Column(String)
    prcp = Column(Float)
    tobs = Column(Float)

class Station(Base):
    __tablename__ = 'station'
    id = Column(Integer, primary_key=True)
    station = Column(String(255))
    name = Column(String(255))
    latitude = Column(Float)
    longitude = Column(Float)
    elevation=Column(Float)
```

```python
Base.metadata.create_all(engine)
```

```python
def populate_table(engine, table, csvfile):
    conn=engine.connect()
    df_of_data_insert=pd.read_csv(csvfile)
    data=df_of_data_insert.to_dict(orient='records')
    print(data)
    conn.execute(table.delete())
    conn.execute(table.insert(),data)
populate_table(engine, Station.__table__, "hawaii_stations.csv")
populate_table(engine, Measurement.__table__, "clean_hawaii_measurements.csv")
```
